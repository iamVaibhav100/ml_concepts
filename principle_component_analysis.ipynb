## Principle Component Analysis (PCA)

- PCA is a technique used to reduce the dimensionality of a dataset by identifying a new set of uncorrelated variables called principal components. It helps to extract meaningful information from high-dimensional datasets and represent it in a reduced number of dimensions, making it easier to analyze and visualize.

- PCA works by finding the directions in the data that capture the most variation. These directions are the eigenvectors of the covariance matrix of the data. The eigenvectors with the largest eigenvalues correspond to the principal components.

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
y = iris.target
```

```python
X[:10]
```

```python
y[:10]
```

Standardize the features using the StandardScaler class from scikit-learn. This will ensure that each feature has zero mean and unit variance, which is necessary for PCA

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
X_std[:10]
```

```python
plt.scatter(X_std[:, 0], X_std[:, 1], c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)
```

```python
X_pca[:10]
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.show()
```
